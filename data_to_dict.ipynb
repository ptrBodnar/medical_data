```python
import pandas as pd
```

```python
df = pd.read_json('data_long.json')
ind_max = pd.read_json('indicators_max.json', typ='series')
```

```python
def get_indicator_d(i):
    ind = i.iloc[0].indicator
    dct = {
        'ind': ind,
        'point2018': i.loc[i.year == 2018, 'value'].iloc[0],
        'point2017': i.loc[i.year == 2017, 'value'].iloc[0],
        'city': i.city.iloc[0]
    }
    
    dct['normPoint2017'] = dct['point2017'] / ind_max[ind] * 100
    dct['normPoint2018'] = dct['point2018'] / ind_max[ind] * 100
    return ind, dct

def nest(d):
    return {
        'city': d.iloc[0].city,
        'district': d.iloc[0].district,
        'lon': d.iloc[0].lon,
        'lat': d.iloc[0].lat,
        'indicators': {
            k: v for k, v in d.groupby('indicator', as_index=False).apply(get_indicator_d).values
        }
        
    }
```

```python
df.groupby('city').apply(nest).to_json('data_nested.json')
```

```python
df.groupby('city').apply(nest).to_dict()
```

```python
def make_xls(d):
    city = d.iloc[0].city
    writer = pd.ExcelWriter(f'../src/assets/data/xls/{city}.xlsx', engine='xlsxwriter')
    
    pivoted = d[['indicator', 'value', 'year']].pivot(index='year', columns='indicator', values='value').T
    pivoted['max'] = [ind_max[i] for i in pivoted.index.values]
    pivoted = pd.concat([pd.DataFrame([[None, None, None]], index=[city], columns=[2017, 2018, 'max']), pivoted])
    
    d2017 = pivoted[[2017, 'max']]
    d2018 = pivoted[[2018, 'max']]
    d2017.columns = ['бал', 'max']
    d2018.columns = ['бал', 'max']
    
    d2017.to_excel(writer, sheet_name='2017')
    d2018.to_excel(writer, sheet_name='2018')
    
    writer.save()

test = df.groupby('city', as_index=False).apply(make_xls)
```
