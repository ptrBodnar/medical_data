```python
import json
import requests
import itertools
import pandas as pd
from pandas.io.json import json_normalize
```

```python
sad_ids = ["25845655", "25843931", "03451526", "25891336", "25825434", "26345736", "24718432", "25946285", "25908960", 
           "25253009", "25878206", "25829550", "25898491", "25894919", "24014538", "25887079", "25449824", "30885376", 
           "25899361", "25879878", "05422303", "22843868", "25901106"]
```

```python
def download_docs(ids, d_type):
    
    docs = []
   
    for i in ids:
        q1 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-01-01&documentDateTo=2018-03-31").json()['documents']
        q2 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-04-01&documentDateTo=2018-06-30").json()['documents']
        q3 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-07-01&documentDateTo=2018-09-30").json()['documents']
        q4 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-10-01&documentDateTo=2018-12-31").json()['documents']
        
        #another year
#         q1 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-01-01&documentDateTo=2018-03-31").json()['documents']
#         q2 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-04-01&documentDateTo=2018-06-30").json()['documents']
#         q3 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-07-01&documentDateTo=2018-09-30").json()['documents']
#         q4 = requests.get("http://api.spending.gov.ua/api/v2/disposers/" + d_type + "?disposerId=" + i + "&documentDateFrom=2018-10-01&documentDateTo=2018-12-31").json()['documents']

        
        docs_year = q1 + q2 + q3 + q4
        docs.append(docs_year)
    
    docs = list(itertools.chain(*docs))
    return(docs)
```

```python
contracts = download_docs(sad_ids, "contracts")
addendums = download_docs(sad_ids, "addendums")
acts = download_docs(sad_ids, "acts")
```

```python
with open('contracts.json', 'w') as outfile:
    json.dump(contracts, outfile, ensure_ascii=False)

with open('addendums.json', 'w') as outfile:
    json.dump(addendums, outfile, ensure_ascii=False)

with open('acts.json', 'w') as outfile:
    json.dump(acts, outfile, ensure_ascii=False)
```

```python
addendums_table = json_normalize(addendums, record_path='contractors', meta=['amount', 'amountIncrease', 'correctionType', 
                                                                       'correctionVATValue', 'currency', 'currencyAmountUAH', 
                                                           'documentDate', 'documentNumber', 'edrpou', 'fromDate', 
                                                           'id', 'isCorrectionWithPDF', 'isCpvVat', 'noTerm', 
                                                           'parentId', 'pdvAmount', 'pdvInclude', 'procurementItems', 
                                                           'reasonOtherComment', 'reasonTypes', 'signDate', 
                                                           'specifications', 'subject', 'toDate'])
pd.DataFrame.to_csv(addendums_table, "addendums.csv")
```

```python
contracts_table = json_normalize(contracts, record_path='contractors', meta=['amount', 'currency', 'currencyAmountUAH',
       'documentDate', 'documentNumber', 'edrpou', 'fromDate', 'id',
       'isCpvVat', 'noTerm', 'pdvAmount', 'pdvInclude', 'procurementItems',
       'reason', 'signDate', 'specifications',
       'subject', 'tender', 'toDate'])

pd.DataFrame.to_csv(contracts_table, "contracts.csv")
```
